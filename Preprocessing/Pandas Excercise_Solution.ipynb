```python
mean_squared_error(actual_age, model_pred_age) #MSE
```

```python
mean_absolute_error(actual_age, model_pred_age) #MAE
```

```python
np.sqrt(mean_squared_error(actual_age, model_pred_age))#RMSE
```

```python
r2_score(actual_age, model_pred_age)
```

```python
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
mean_absolute_percentage_error(actual_age, model_pred_age)
```
